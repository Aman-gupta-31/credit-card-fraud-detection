## Dataset Information

This dataset chronicles 284,807 credit card transactions executed by European cardholders across two days in September 2013. A critical characteristic is its severe imbalance, as only 492 instances—a mere 0.172% of the total—are classified as fraud (the positive class).

The data comprises numerical input variables. Specifically, V1 through V28 are features derived from a Principal Component Analysis (PCA) transformation, implemented due to confidentiality restrictions, meaning the original feature details are unavailable. The remaining two features, 'Time' and 'Amount', have not been transformed by PCA. The 'Time' feature records the seconds that have elapsed since the dataset's first transaction, while the 'Amount' feature contains the transaction value itself, making it relevant for example-dependent cost-sensitive learning. The output variable, 'Class', indicates the transaction status, taking a value of 1 for fraud and 0 otherwise.

Given the substantial disparity in class distribution, relying on traditional Confusion Matrix Accuracy for evaluation is not advisable, as it yields misleading results for highly unbalanced classification problems. Instead, we strongly recommend using the Area Under the Precision-Recall Curve (AUPRC) as the appropriate measure for model performance assessment.

## Input Split

```python
X = df.drop(columns=['Class'], axis=1)
y = df['Class']

# Drop rows where the target variable 'Class' is NaN
df.dropna(subset=['Class'], inplace=True)
X = df.drop(columns=['Class'], axis=1)
y = df['Class']
```

## Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaler = sc.fit_transform(X)
```

```python
x_scaler[-1]
```

## Model Training

```python
# train test split
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score
x_train, x_test, y_train, y_test = train_test_split(x_scaler, y, test_size=0.25, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(n_jobs=-1)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

## Class Imbalancement

```python
print(y_train.value_counts())
```

```python
# balance the class with equal distribution
from imblearn.over_sampling import SMOTE
over_sample = SMOTE()
x_smote, y_smote = over_sample.fit_resample(x_train, y_train)
```

```python
print(y_smote.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_smote, y_smote)

y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_jobs=-1)
model.fit(x_smote, y_smote)

y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(n_jobs=-1)
model.fit(x_smote, y_smote)

y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```
